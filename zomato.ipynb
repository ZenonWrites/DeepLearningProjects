```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("E:\csv\zomato.csv")
df.head()
```

```python
df = df.drop(['url', 'address', 'phone','menu_item','reviews_list','dish_liked'],axis = 1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if( value == 'NEW' or value == '-'):
     return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)
df.rate.isnull().sum()
```

```python
df.info()
```

```python
df.info()
```

```python
df.dropna(inplace =True)
df.head()
```

```python
df.info()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df = df.drop(['listed_in(city)'],axis = 1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

```python
def handlercomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlercomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type = df['rest_type'].value_counts(ascending =  False)
rest_type
```

```python
rest_type_lessthan1000 = rest_type[rest_type < 1000]
rest_type_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan1000):
      return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)

df['rest_type'].value_counts()

```

```python
df.head()

```

```python
location_type = df['location'].value_counts(ascending = False)
location_type
```

```python
location_type_lessthan1000 = location_type[location_type < 300]
location_type_lessthan1000 
```

```python
location_type.value_counts()
```

```python
df['location'].value_counts()
```

```python
def handle_location(value):
    if(value in location_type_lessthan1000 ):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
